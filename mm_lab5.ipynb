```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
```

#Радиоактивный распад

```python
k = np.log(2) / 1600  # Коэффициент распада для полураспада 1600 лет
x0 = 1  # Начальное количество вещества

# Модель радиоактивного распада
def radioactive_decay(x, t):
    return -k * x

# Временной интервал
t = np.linspace(0, 100, 50)

# Численное решение
x_numeric = odeint(radioactive_decay, x0, t)

# Аналитическое решение
x_analytic = x0 * np.exp(-k * t)

# График
plt.plot(t, x_numeric, color='c', label='Численное решение')
plt.plot(t, x_analytic, '+', color='m', label='Аналитическое решение')
plt.xlabel('Время (лет)')
plt.ylabel('Количество вещества')
plt.title('Модель радиоактивного распада')
plt.legend()
plt.grid(True)
plt.show()
```

# Охлаждение

```python
T0 = 100  # Начальная температура тела
Ts = 20   # Температура окружающей среды
k = 0.1   # Константа охлаждения

# Модель охлаждения
def cooling(T, t):
    return -k * (T - Ts)

# Временной интервал
t = np.linspace(0, 10, 50)

# Численное решение
T_numeric = odeint(cooling, T0, t)

# Аналитическое решение
T_analytic = Ts + (T0 - Ts) * np.exp(-k * t)

# График
plt.plot(t, T_numeric, color='c', label='Численное решение')
plt.plot(t, T_analytic, '+', color='m', label='Аналитическое решение')
plt.xlabel('Время (минуты)')
plt.ylabel('Температура')
plt.title('Модель охлаждения тела')
plt.legend()
plt.grid(True)
plt.show()
```

# Реакция на рекламу

```python
N0 = 1000  # Общее число людей
x0 = 10    # Количество людей, знающих о продукте в начале
k = 0.02   # Коэффициент распространения

# Модель реакции на рекламу
def advertising(x, t):
    return k * x * (N0 - x)

# Временной интервал
t = np.linspace(0, 10, 50)

# Численное решение
x_numeric = odeint(advertising, x0, t)

# Аналитическое решение (логистическая кривая)
A = N0 / x0 - 1
x_analytic = N0 / (1 + A * np.exp(-k * N0 * t))

# График
plt.plot(t, x_numeric, color='c', label='Численное решение')
plt.plot(t, x_analytic, '+', color='m', label='Аналитическое решение')
plt.xlabel('Время (дни)')
plt.ylabel('Число знающих о продукте')
plt.title('Модель реакции на рекламу')
plt.legend()
plt.grid(True)
plt.show()
```

# Математический маятник

```python
# Параметры маятника
g = 9.81  # Ускорение свободного падения, м/с^2
L = 1.0   # Длина маятника, м

# Начальные условия: [начальный угол (рад), начальная угловая скорость (рад/с)]
theta0 = [np.pi / 4, 0]

# Модель математического маятника
def pendulum(theta, t, g, L):
    dtheta_dt = theta[1]
    domega_dt = -(g / L) * np.sin(theta[0])
    return [dtheta_dt, domega_dt]

# Временной интервал
t = np.linspace(0, 10, 500)

# Численное решение
theta_numeric = odeint(pendulum, theta0, t, args=(g, L))

# Аналитическое решение для малых углов
theta_analytic = theta0[0] * np.cos(np.sqrt(g / L) * t)

# График
plt.plot(t, theta_numeric[:, 0], color='c', label='Численное решение')
plt.plot(t, theta_analytic, '+', color='m', label='Аналитическое решение (малые углы)')
plt.xlabel('Время (с)')
plt.ylabel('Угол (рад)')
plt.title('Модель математического маятника')
plt.legend()
plt.grid(True)
plt.show()
```
